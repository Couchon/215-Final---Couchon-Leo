### Introduction ###

       On September 9 to the early morning of the 10th on the Oregon Coast in 2022, there were hot and dry Easterly winds paired with a coastal fire. These winds caused the Cedar Creek fire to uproar and push its smoke offshore. The Regional Cabled Array rested beneath the water monitoring the smokescreen and enabling impact analysis of the smoke. This enabled a glimpse into the interdisciplinary study of terrestrial disasters on marine life. 
        Through sailing with the OOI during the summer of â€˜24, I was able to witness firsthand the real time problem solving of engineers and scientists as they updated equipment and mitigated the effects of biofouling at the Oregon Offshore Site. The water there is rich with life and poses a consistent maintenance issue for the people who service the delicate instruments along the fiber optic cable. I was absorbed and filled with questions on how dynamic such a rich environment is and was particularly interested in the lower trophic levels that maintain the observed ecosystem. One of the pieces of instruments I found myself enamored by was the EK60 that was being replaced by the EK80 that cruise. An engineer informed me it was biosonar equipment that could pick up creatures as small as zooplankton and that its data is largely underutilized. Prior to this, a researcher at the University of Washington, Dr. Wu-Jun Lee, had set out to make this data approachable and looked at the eclipse's effect on the zooplankton. The zooplankton were tricked by the eclipse and came to the surface only to have to return back down at the eclipse's end.Another study, 'Observations of Phytoplankton Community Composition in the Santa Barbara Channel During the Thomas Fire' by Sasha Kramer, Kelsey Bisson, and Alexis Fescher also spiked my curiosity as it gave me insight into how the affiliated deposits of a smokescreen can impact the RedField Ratio and can cause an iron increase resulting in phytoplankton blooms. This was something I would like to continue looking into in the future but was unfortunately unable to explore with this project as many instruments were down at the time of historic wildfire offshore pushes: fluorometry data was unavailable and I was limited to one biosonar site during the time of the Cedar Creek fire. 
        Dr Wu-Jun Lee's open source python and text explaining her analytical processes was instrumental in my understanding of the data manipulation and results identification. The researchers who analyzed the Santa Barbara fires were also important for me to be able to see how far out the reaches of a single impact can lay. # 1. Echograms vis EK60 Bio Sonar #
Week of 8/28 - 10/02 of bioacoustic sonar frequencies over ping time in UTC time and depth in meters. The vertical range was sorted into bins for meter depth prior to processing. The ideal frequency for visibility is around 200 kHz and was cleaned so for this purpose. 

```python
#filepaths
fp_828_904 = 'calib_ek60_nc_mj01c/CE02SHBP_Bioacoustic_Echogram_20220828-20220904_Calibrated_Sv_Averaged.nc'
fp_904_911 = 'calib_ek60_nc_mj01c/CE02SHBP_Bioacoustic_Echogram_20220904-20220911_Calibrated_Sv_Averaged (1).nc'
fp_911_918 = 'calib_ek60_nc_mj01c/CE02SHBP_Bioacoustic_Echogram_20220911-20220918_Calibrated_Sv_Averaged.nc'
fp_918_925 = 'calib_ek60_nc_mj01c/CE02SHBP_Bioacoustic_Echogram_20220918-20220925_Calibrated_Sv_Averaged.nc'
fp_925_1002 = 'calib_ek60_nc_mj01c/CE02SHBP_Bioacoustic_Echogram_20220925-20221002_Calibrated_Sv_Averaged.nc'

#open data
xr_828_904 = xr.open_dataset(fp_828_904)
xr_904_911 = xr.open_dataset(fp_904_911)
xr_911_918 = xr.open_dataset(fp_911_918)
xr_918_925 = xr.open_dataset(fp_918_925)
xr_925_1002 = xr.open_dataset(fp_925_1002)

#clean data to desired frequency
xr_828_904_200 = xr_828_904.sel(frequency_nominal=200000)
xr_904_911_200 = xr_904_911.sel(frequency_nominal=200000)
xr_911_918_200 = xr_911_918.sel(frequency_nominal=200000)
xr_918_925_200 = xr_918_925.sel(frequency_nominal=200000)
xr_925_1002_200 = xr_925_1002.sel(frequency_nominal=200000)


#define variables within data
#8/28-9/04/2022
times_828_904 = xr_828_904_200['ping_time']
depth_828_904 = xr_828_904_200['nominal_depth']
sv_828_904 = xr_828_904_200['Sv']

#9/04-9/11/2022
times_904_911 = xr_904_911_200['ping_time']
depth_904_911 = xr_904_911_200['nominal_depth']
sv_904_911 = xr_904_911_200['Sv']

#9/11-9/18/2022
times_911_918 = xr_911_918_200['ping_time']
depth_911_918 = xr_911_918_200['nominal_depth']
sv_911_918 = xr_911_918_200['Sv']

#9/18-9/25/2022
times_918_925 = xr_918_925_200['ping_time']
depth_918_925 = xr_918_925_200['nominal_depth']
sv_918_925 = xr_918_925_200['Sv']

#9/25-10/02/2022
times_925_1002 = xr_925_1002_200['ping_time']
depth_925_1002 = xr_925_1002_200['nominal_depth']
sv_925_1002 = xr_925_1002_200['Sv']
```

```python
#Map the Echograms
fig, axs = plt.subplots(5, 1, figsize=(30, 20))

echogram1 = axs[0].pcolormesh(times_828_904, depth_828_904, sv_828_904.T, cmap='RdBu_r')
axs[0].set_xlabel('Time (UTC)', fontsize =20)
axs[0].set_ylabel('Depth (m)', fontsize =20)
axs[0].set_title('Zooplankton Biosonar Echogram-8/28-9/04/2022', fontsize =25)
cb1 = plt.colorbar(echogram1)
#Removed label for half of the echograms for better legibility since all the same label='Strength Volumbe Backscatter (dB)'
echogram2 = axs[1].pcolormesh(times_904_911, depth_904_911, sv_904_911.T, cmap='RdBu_r')
axs[1].set_xlabel('Time (UTC)', fontsize =20)
axs[1].set_ylabel('Depth (m)', fontsize =20)
axs[1].set_title('Zooplankton Biosonar Echogram-9/04-9/11/2022', fontsize =25)
cb2 = plt.colorbar(echogram2, label='Strength Volumbe Backscatter (dB)')

echogram3 = axs[2].pcolormesh(times_911_918, depth_911_918, sv_911_918.T, cmap='RdBu_r')
axs[2].set_xlabel('Time (UTC)', fontsize =20)
axs[2].set_ylabel('Depth (m)', fontsize =20)
axs[2].set_title('Zooplankton Biosonar Echogram-9/11-9/18/2022', fontsize =25)
cb3 = plt.colorbar(echogram3) #, label='Strength Volumbe Backscatter (dB)'

echogram4 = axs[3].pcolormesh(times_918_925, depth_918_925, sv_918_925.T, cmap='RdBu_r')
axs[3].set_xlabel('Time (UTC)', fontsize =20)
axs[3].set_ylabel('Depth (m)', fontsize =20)
axs[3].set_title('Zooplankton Biosonar Echogram-9/18-9/25/2022', fontsize =25)
cb4 = plt.colorbar(echogram4, label='Strength Volumbe Backscatter (dB)')

echogram5 = axs[4].pcolormesh(times_925_1002, depth_925_1002, sv_925_1002.T, cmap='RdBu_r')
axs[4].set_xlabel('Time (UTC)', fontsize =20)
axs[4].set_ylabel('Depth (m)', fontsize =20)
axs[4].set_title('Zooplankton Biosonar Echogram-9/25-10/02/2022', fontsize =25)
cb5 = plt.colorbar(echogram5) #, label='Strength Volumbe Backscatter (dB)'

#Make Plots Look Good
plt.tight_layout()
```

*Figure Caption : The white up and down motion can be followed till the day of smoke on the ninth. A strong behavioural shift can be observed as the zooplankton seem to spend more time in the mid water column with inconstiency of day night pattern. ## Limitations of Research and Possible Next Steps ## 
### Failures So Far ###
1. Can not concretly prove redfield ratio change without in situ sampling. (Second hypothesis is near impossible to prove with current data sets, would need to expand)

2. Can not sample for community changes (what is blooming is it a new community is it one that thrives slighlty deeper in the water column?)

3. Minimal data cleaning to see the full picture simultaneously gives us little to concretely say about what we are looking at - (such as - what is equipment malfunction, what is noise, what is debris)

4. Was not able to find any information on how equipment used at RCA equipment from the Applied Physics Laboratories handles 

#### Possible Future Solutions in next steps ###
1. Test equipment in water that is calibrated for an environment that has been introduced to similar conditions as water in wildfires - how does equipment behave with debris? Sudden change in chemical parameters? Sudden change in biofouling from said changes? Could this be a possible reason for multiple parameters having an odd drift?

2. Increase data set timeline - look at multiple more seasons, a whole year or/and look at multiple reoccurences of the same timeline (a few more years of August, September and October data to understand change) This would be difficult at current site due to huge data holes. 

3. Set up biosonars close to the coast in fireprone areas that are more accessible and can then be serviced more regularly

4. Have scientists that can essentially 'fire chase' for sampling during fire season as sites with EK60's and EK80's

5. An exciting next hypothesis question could be analyzing how the Redfield Ratio at these given sites changes, how intensity of wildfires and concentration of smokescreen impacts the data and how the community changes and from before during and after wildfire smokes in both the zooplanlton and phytoplankton communities (how do they respond to changes in each other, which is impacted more immediately? To what degree? How does the chemcial compositon of the water change with this? 

6. The possibilited of where to take this research are endless since we are looking at a major ecosystem variable! We could also look at how it impacts large mammals, large mammal migration patterns, oxygenation of the water column, biofouling and more.
